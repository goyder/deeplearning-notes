# Convolutional Neural Nets

## Convolutions and shape reduction

Applying a convolution without adding any padding will cause the output image to "shrink".

Consider an $n \times n$ matrix, and an $f \times f$ filter.

The output size of the filtered (or convolved? Convoluted?) matrix will be $n-f+1 \times n-f+1$.

To counteract this, we add the use of padding, where we join typically 0-pixels on the border of the image. 

If we perform a convolution, but add $p$ padded pixels to the outside of the image, the output shape is $n-f+1+2p \times n-f+1+2p.$



## Strided convolutions

Rather than taking the convolution of every possible centre point on the matrix, we might choose to jump - or "step" - over some of the components in the dataset.

We therefore add a new variable - stride, or $s$. Thus for a given stride, the output shape becomes $\frac{n+2p-f}{s}+1 \times \frac{n+2p-f}{s}+1$.

## A practical attempt at convolutions

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = 10, 10
%matplotlib inline

from scipy.signal import convolve2d
from scipy.ndimage import imread
```

```python
bart = imread("static/bart 2.png", mode="L")
scary = imread("static/scary.jpg", mode="L")
```

```python
plt.imshow(bart)
```

```python
convolution = np.array([
    [0, 1, -1, 0],
    [1, 3, -3, -1],
    [1, 3, -3, -1],
    [0, 1, -1, 0]
])
```

```python
plt.imshow(convolve2d(bart, convolution))
```

```python
plt.figure(figsize=(15,15))
plt.imshow(convolve2d(scary, convolution))
```
